## Building of a Recuurenet Neural Network(RNN)....

```python
import keras
#for initialising the RNN i am import Sequential models from kears.models....
from keras.models import Sequential
#for adding output layers i am importing the Dense layer from yhe keras.layers...
from keras.layers import Dense
#for Vanishing the Gradient Descent  i am going to use Long Short Term Memory Network(LSTM)
from keras.layers import LSTM
#for regularisation and to overcoming on the overfitting i am going to import Dropout layer...
from keras.layers import Dropout
```

```python
#Initializing the RNN.....
regressor=Sequential()
#here i am using the name of my  model i sregressor because RNN is a regression model
# if we are predicting a continuous value then that type of model is a regression model..
#while if we are predicting a categorical value then taht6 type of model is called a classifier model....that are ANN and CNN
```

```python
# ddding the first LSTM layer and some Dropout layer...
# adding 1st layer
regressor.add(LSTM(units=50,return_sequences=True,input_shape=(60,1)))
# units can be any random number it not neccessary to take 50 always but in my case 50 is a appropriate number..
#return_sequences=True its because i want to repeat it again with LSTM layer....
# input_shape must be same shape as that of training shjape except first one which is the total number of data in a column..
regressor.add(Dropout(0.2))
# Dropout=0.2 menas i am ignoring 20% of neurons to overcome on overfitting and 0.2 is most appropriate in most case.
# adding 2nd layer
regressor.add(LSTM(units=50,return_sequences=True))
regressor.add(Dropout(0.2))
 # adding 3rd layer
regressor.add(LSTM(units=50,return_sequences=True))
regressor.add(Dropout(0.2))
# adding 4th layer and it is last layer...
## return_sequences will  by default False that's why i am not  writing return_sequences=False.. 
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))
```

```python
# Adding the final layer that is the output layer...using the Dense layer...
regressor.add(Dense(units=1))
# here the uniits is equal to one because we need only one output layer for getting the predictons..
```

```python
# Comiling the RNN model....
regressor.compile(optimizer='rmsprop',loss='mean_squared_error',metrics=['accuracy'])
# For optimizer we can also use 'adam' optimization method but for a RNN model rmsprop optimization method is very efficient...
# Here i am using mean_squared_erroir beacuse it is not a classification problem it is a regression problem...
```

```python
# Fitting the RNN to Training_Set..
regressor.fit(x_train,y_train,batch_size=32,epochs=100)
# batch_size is the baunch of the neurons which will take part in the fitting together.....
# Here i am taking the batch_size is 32 is appropriate for my case.... but it is not neccassry i can take any random number..
# epochs is representing the iterations that 100 times. it will calculate and will try to reduce the loss and will find the accuracy....
```
